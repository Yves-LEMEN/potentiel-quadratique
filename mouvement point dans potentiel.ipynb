# Mouvement d'un point dans un potentiel quadratique

***

## Introduction

Un mobile de masse $m$ glisse le long d'un ﬁl dont l’équation est $y=f(x)$. 
Il n’est soumis qu’à la pesanteur. 

Le potentiel dû à la gravité est donc $mg[f(x)-f(x_0)]$

On définit l'origine en sorte que $f(x_0)=0$ et on suppose que la fonction $f$ présente un minimum en $x_0$. On peut alors développer la fonction $f$ en série au voisinage de zéro, pour définir le potentiel:

$$
\mathrm{V}(x) = \frac{mg}{2}\ddot{f}(0)x^2 + \mathrm{O}(x^3)\\
 = \frac{1}{2}m\omega^2x^2,
$$

En posant $\omega = \sqrt{\frac{g}{m}\ddot{f}(0)}$


Le Lagrangien s'écrit :
$$
\mathrm{L}(x, \dot{x}) = \frac{1}{2}m \dot{x}^2-\frac{1}{2}m \omega^2 s^2
$$
Et on aura pour l'équation de Lagrange  $$\frac{d}{dt}[\frac{\partial\mathrm{L}}{\partial\dot{x}}]=\frac{\partial\mathrm{L}}{\partial{x}}\implies \boxed{ \ddot{x}= -\omega^2 x}
$$


Equation dont la solution est de la forme $A sin(\omega t + \phi)$ mais que l'on peut résoudre numériquement.




## La méthode "leapfrog"

Pour éviter d'avoir' à résoudre une équation du second ordre, on va plutôt résoudre un système de deux équations du premier ordre :

\begin{eqnarray}
\frac{dx}{dt} & = & v \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; (1) \\
\frac{dv}{dt} & = & -\omega^2x, \;\;\;\;\;\;\;\;\;\;\; (2)
\end{eqnarray}

Le principe de l'intégration "saute-mouton" (leapfrog en Anglais) est de travailler alternativement sur la suite $x$ et la suite $v$. On transforme donc les système (1) (2) en :

\begin{eqnarray}
\frac{x_{n+1}-x_{n}}{\Delta t} & = & v_{n+1} \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; (3) \\
\frac{v_{n+1}-v_{n}}{\Delta t} & = & -\omega^2x_n, \;\;\;\;\;\;\;\;\;\;\; \;(4)
\end{eqnarray}

Soit encore :
\begin{eqnarray}
x_{n+1} & = & x_{n}+v_{n+1}\Delta t \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; (5) \\
v_{n+1} & = & v_{n} -\omega^2 \Delta t x_n, \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;(6)
\end{eqnarray}


```python
# Résolution de l'équation différentielle du second degré y"=-ky par la méthode leapfrog


from pylab import *
import matplotlib.pyplot as plt
###############################################################################
# définition du pas d'intégration N
N = 1000
tps=8.0
t = linspace(0,tps,N)
dt = t[1] - t[0]

###############################################################################
# création de la fonction integrer
def resoudre(x0,v0):
    ###########################################################################
    # Initialisation à 0 des variables, et Conditions initiales
    x = zeros(N)
    v = zeros(N)
    x[0] = x0
    v[0] = v0
    
    ###########################################################################
    # Intégration "leapfrog"
        
    for i in range(N-1):
         v[i + 1] = v[i] - dt*x[i]
         x[i + 1] = x[i] + dt*v[i + 1]
           
    ###########################################################################
    # Solution
    return x,v

###############################################################################
# retour au programme principal et assignation des quantités x0 & v0 (conditions initiales)
x,v = resoudre(-20,-1.5) 

###############################################################################
# représentation graphique
rcParams["axes.grid"] = True
rcParams['font.size'] = 14
rcParams['axes.labelsize'] = 20
figure()
plt.plot(t,x,'g',linewidth=0.5)
plt.plot(t,v,'r',linewidth=0.5)
xlabel("temps")
ylabel("x(t), v(t)")

###############################################################################
show()
###############################################################################
```



## "temps de propagation" et "pas d'intégration"

On trouve effectivement des courbes présentant une allure en $Asin(\omega t + \phi)$ 

On doit remarquer que les équations du mouvement écrites sous leur forme (3)-(4) ci-dessus ne permettent un calcul effectif que dans la mesure où $\Delta t$ est petit, sachant que l'on a:

$$
\frac{v^2}{v_0^2} + \frac{x^2}{x_0^2}=1 \\\\\
v_0 = \sqrt{\frac{2E}{m}}   \quad  \;\;    x_0=\frac{1}{\omega }\sqrt{\frac{2E}{m}}\\
$$
Dans les approximations faites lorsqu'on remplace la fonction par une suite de segments, on a $\Delta t \approx \large{\frac{2 tps}{N}}$ où  1/N désigne le pas d'intégration et tps le temps d'intégration, comme ci-dessus dans l'instruction "t=linspace(0,tps,N)".

Si veut que les égalités continues (1) et (2) soient bien approchées par les suites discrètes (3) (4), l'incrément sur chaque suite doit être petit, donc en particulier:
- pour les positions $\large{\frac{2 tps}{N}} \ll1$
- pour les vitesses $\mathrm \omega=\large{\frac{v_0}{x_0}}$ doit être tel que l'incrément $\omega\Delta t \ll1$ c'est à dire $\large{\frac{v_0}{x_0}}\Delta t \ll1$
 
Par ailleurs, on voit qu'on a $\omega\;\# \;2\pi /tps$ donc, $tps/N<1/6$ si on veut effectuer une intégration sur plus d'une unité de temps. Cette seconde exigence peut être plus sévère que  celle realtive aux positions qui est de même nature.




```python
# retour au programme principal et assignation des quantités x0 & v0 (conditions initiales)
N = 40
tps=8.0
t = linspace(0,tps,N)
dt = t[1] - t[0]

x,v = resoudre(0.5,1.0) 

###############################################################################
# représentation graphique
rcParams["axes.grid"] = True
rcParams['font.size'] = 14
rcParams['axes.labelsize'] = 20
figure()
plt.plot(t,x,'g',linewidth=0.5)
plt.plot(t,v,'r',linewidth=0.5)
xlabel("temps")
ylabel("x(t), v(t)")

###############################################################################
show()
###############################################################################
```
